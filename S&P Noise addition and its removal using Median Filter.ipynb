# Problem Statement: 
#### Read a color image. Add salt and pepper noise to it. Use a 3 x 3 median filter to remove the noise.

### Importing the required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Taking a color image as input 

```python
img=cv2.imread('ComputerVision/DATA/dog_backpack.jpg')
```

```python
type(img)
```

### Printing the image taken as input


```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(18,12))
plt.imshow(img)
```

```python
img.shape
```

### Adding Salt and Pepper Noise to the image

```python
# Salt and Pepper noise cannot be added to color image. Therefore we have to convert the given image to grayscale.
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(18,12))
plt.imshow(img,cmap='gray')
```

```python
img.shape
```

```python
noisy_img=img.copy()
```

#### Adding Salt Noise

```python
n_salt=np.random.randint(50000,100000)
```

```python
for i in range(0,n_salt):
    x=np.random.randint(0,img.shape[0])
    y=np.random.randint(0,img.shape[1])
    noisy_img[x,y]=255
```

```python
plt.figure(figsize=(18,12))
plt.imshow(noisy_img,cmap='gray')
```

#### Adding Pepper Noise

```python
n_pepper=np.random.randint(50000,100000)
```

```python
for i in range(0,n_pepper):
    x=np.random.randint(0,img.shape[0])
    y=np.random.randint(0,img.shape[1])
    noisy_img[x,y]=0
```

```python
plt.figure(figsize=(18,12))
plt.imshow(noisy_img,cmap='gray')
```

### Removing Salt and Pepper Noise using 3X3 Median Filter

```python
medFil_img=noisy_img.copy()
```

```python
medFil_img=medFil_img.astype(np.uint8)
```

```python
medFil_img=cv2.medianBlur(medFil_img,3)
```

```python
plt.figure(figsize=(18,12))
plt.imshow(medFil_img,cmap='gray')
```

### Displaying Final Results

```python
fig = plt.figure(figsize=(18,12))
ax1 = fig.add_subplot(131)
ax1.set_title('Original Image')
ax1.imshow(img,cmap='gray')
ax2 = fig.add_subplot(132)
ax2.set_title('Image with Salt & Pepper Noise')
ax2.imshow(noisy_img,cmap='gray')
ax3 = fig.add_subplot(133)
ax3.set_title('Image after applying Median Filter')
ax3.imshow(medFil_img,cmap='gray')
```
